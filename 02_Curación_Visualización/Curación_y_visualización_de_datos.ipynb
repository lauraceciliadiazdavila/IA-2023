## EDA

El analisis exploratorio de datos o EDA (Exploratory data analysis) nos permite organizar y entender las variables para asi poder identificar los diferentes tipos de las mismas, las categorías a la que pertenecen y el tipo de análisis que se puede realizar sobre ellas. Tambien se pueden establecer relaciones, encontrar patrones ocultos en los datos, lo que permitira encontrar información o comportamientos relevantes.

Ayuda a escoger el modelo correcto para la necesidad correcta, gracias al entendimiento que obtenemos al realizar una EDA, como asi tambien te ayuda a tomar decisiones informadas, basadas en los datos, sus relaciones, etc.

La idea es hacer preguntas sobe los datos para guiar la EDA como por ejemplo:
* Que te gustaria encontrar?
* Que quisieras saber de los datos?
* Cual es la razon para realizar el analisis?
* Cuantas observaciones existen?
* Cuantas variables hay?
* Necesito todas las observaciones?
* Necesito todas las variables?  
* Cuantas variables categóricas existen?
* Cuantas variables continuas existen?
* Como puedo explorar cada variable dependiendo de su categoría?

Una vez realizada la EDA y teniendo conocieminto, pasamos a la limpieza en donde las preguntas que nos podrian guiar serian:
* Tengo valores faltantes?
* Cual es la proporción de datos faltantes?
* Como puedo tratar a los datos faltantes?
* Cual es la distribución de los datos?
* Tengo valores atipicos?

Luego de la limpieza, establecer relaciones entre los datos me ayudaria a responder preguntas como:

* Existe algun tipo de relacion entre mi variable X y Y?
* Que pasa ahora si considero la variable Z en el analisis?
* Que significa que las observaciones se agrupen?
* Que significa el patron que se observa?

> Este proceso es ciclico. A pesar de que pueda parecer infinito, este proceso en algun momento debe salir del ciclo y continuar para obtener algun tipo de valor

## Matplotlib

---

**Grafico de torta**

---

```python
#Datos
x = df["species"].value_counts()
labels = df["species"].values.unique()
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, labels=labels ,colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)
ax.set_title("Porcentaje de cada especie")

ax.set(xlim=(0, 8), xticks=np.arange(1, 10),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
