##Transformers de Hugging Face

¿Qué es la librería "transformers"?
Es una librería de software desarrollada por Hugging Face que proporciona una interfaz fácil de usar para trabajar con modelos de lenguaje basados en la arquitectura Transformer. Estos modelos son capaces de realizar tareas como traducción automática, generación de texto, análisis de sentimientos, y más.

¿Qué hace la librería "transformers"?

Carga y utiliza modelos de lenguaje pre-entrenados, como BERT, GPT, RoBERTa, etc.
Ofrece herramientas para tokenizar texto, es decir, convertir texto en unidades comprensibles para los modelos.
Facilita la generación de texto y la predicción de secuencias.
Permite el fine-tuning de modelos para adaptarlos a tareas específicas de PLN, como clasificación de texto o extracción de información.
¿Por qué es importante?

Simplifica el acceso y uso de modelos de lenguaje avanzados, permitiendo a desarrolladores y científicos de datos trabajar de manera eficiente en proyectos de PLN.
Proporciona una amplia variedad de modelos pre-entrenados, lo que reduce la necesidad de entrenar modelos desde cero en muchas aplicaciones.
Es una herramienta clave en el desarrollo de aplicaciones de PLN, como chatbots inteligentes, sistemas de recomendación de texto, análisis de opiniones, entre otros.
¿Cómo se usa?

Se instala a través de pip (pip install transformers) en entornos de Python.
Se cargan modelos y tokenizadores pre-entrenados utilizando métodos simples de la librería.
Se utilizan pipelines para realizar tareas específicas, como generación de texto o clasificación de texto, de manera fácil y rápida.


```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import ipywidgets as widgets
from IPython.display import display, Markdown

# Configuración inicial
model_id = "gpt2" #https://huggingface.co/openai-community/gpt2
# model_id = "EleutherAI/gpt-j-6B" #https://huggingface.co/EleutherAI/gpt-j-6b
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100)

#text-classification
#sentiment-analysis
```

```python
# Función para generar respuestas utilizando la plantilla y el modelo
def generate_response(question):
    # Generar el prompt a partir de la pregunta
    prompt = format_prompt(question)

    # Usar el pipeline de HuggingFace para generar la respuesta
    response = pipe(prompt)

    # Extraer y devolver el texto generado
    return response[0]["generated_text"]
```

```python
def format_prompt(question):
    # Definir una plantilla básica para el prompt
    template = "Pregunta: {} \n Respuesta:"
    # Combinar la pregunta con la plantilla para crear el prompt final
    prompt = template.format(question)
    return prompt
```

```python
question = 'how are you?'
response = generate_response(question)
display(Markdown(f"### Pregunta:\n{question}\n\n### Respuesta Generada:\n{response}"))
```

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-en-es"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

translation_pipeline = pipeline("translation_en_to_es", model=model, tokenizer=tokenizer)
```

```python
spanish_translation = translation_pipeline(response)
print(spanish_translation[0]["translation_text"])
```

## HuggingFaceEndpoint con LangChain
#Bloque 1: Configuración del entorno y obtención del token


```python
import os
from google.colab import userdata

# Obtener el token de Hugging Face desde los datos de usuario
sec_key = userdata.get("HF_TOKEN")

# Configurar la variable de entorno con el token
os.environ["HUGGINGFACEHUB_API_TOKEN"] = sec_key

# Verificar el token (solo para depuración, elimina esto en producción)
print(sec_key)
```

#Bloque 2: Configuración del modelo de Hugging Face

```python
from langchain_huggingface import HuggingFaceEndpoint

# Definir el ID del repositorio del modelo y otros parámetros
repo_id = "mistralai/Mistral-7B-Instruct-v0.2"
max_length = 128
temperature = 0.7

# Crear la instancia del modelo con Hugging Face Endpoint
llm = HuggingFaceEndpoint(
    repo_id=repo_id,
    max_length=max_length,
    temperature=temperature,
    huggingfacehub_api_token=sec_key
)

# Verificar la conexión y los parámetros del modelo
print("Modelo configurado:", repo_id)
```

#Bloque 3: Definición de la plantilla de prompts

```python
from langchain_core.prompts import PromptTemplate

# Definir la plantilla de prompts
template = """
Question: {question}
Answer: Let's think step by step.
"""

# Crear la instancia de PromptTemplate
prompt_template = PromptTemplate(template=template)
```

#Bloque 4: Definición de la función para generar respuestas

```python
# Función para generar respuestas utilizando la plantilla y el modelo
def generate_response(question):
    # Generar el prompt a partir de la pregunta
    prompt = prompt_template.format(question=question)

    # Usar el endpoint de Hugging Face para generar la respuesta
    response = llm.invoke(prompt)

    # Devolver el texto generado
    return response
```

```python
from IPython.display import Markdown, display

# Hardcodear la pregunta
question = "What is the capital of France?"

# Generar y mostrar la respuesta
response = generate_response(question)
display(Markdown(f"### Pregunta:\n{question}\n\n### Respuesta Generada:\n{response}"))
```
